# Content Based Recommendation System

So we will do content based recommendation based on overview of the movie.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

```python
# Fit he datafrmae on the tfv.This will give sparse matrix.
tfv_data_matrix = tfv.fit_transform(df['overview'])
```

```python
tfv_data_matrix
```

```python
#We will use sigmoid kernel here 

from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
# Compute the sigmoid kernel
sigm_data = sigmoid_kernel(tfv_data_matrix, tfv_data_matrix)
```

```python
sigm_data[5]
```

```python
#we will perform Reverse mapping here to map movie
indices = pd.Series(movies_data.index, index=movies_data['original_title']).drop_duplicates()
```

```python
def find_similarMovies(title, sig=sigm_data):
    # Find  index wrt original_title
    id = indices[title]

    # find  the pairwsie similarity score of the movies 
    sim_scores = list(enumerate(sigm_data[id]))

    # Sort the movies 
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 5 most similar movies
    sim_scores = sim_scores[1:6]

    # Movie indices
    movie_index = [i[0] for i in sim_scores]

    # Top 5 most similar movies
    return movies_data['original_title'].iloc[movie_index]
```

```python
# Lets Test our Content based Recommendation model by providing the movie name
find_similarMovies('Avatar')
```
